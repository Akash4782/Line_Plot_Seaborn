Line Plot

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

x_1 = pd.DataFrame({"var":var , "var_1":var_1})
sns.lineplot(x="var",y="var_1", data = x_1)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins").head(50)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 )
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex" , size =10)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex" , style = "sex" , dashes = False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex" , style = "sex" , palette = "Accent" , dashes = False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex" , style = "sex" , palette = "Accent" , markers = ["o",">"] ,  dashes = False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex" , style = "sex" , palette = "Accent" ,
             markers = ["o",">"] ,  dashes = False , legend = False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data = y_1 , hue = "sex" , style = "sex" , palette = "Accent" ,
             markers = ["o",">"] ,  dashes = False )
plt.grid()
plt.title("Python" , fontsize =20)
plt.show()
```
